```python
from collections.abc import Mapping, Iterable

def deep_min(data):
    if isinstance(data, Mapping):
        return min(deep_min(v) for v in data.values())
    elif isinstance(data, Iterable):
        return min(deep_min(v) for v in data)
    else:
        return data


def main() -> None:
    #data = 5
    #data = [1, [2, 3,[4, 5, 0]]]
    data = {'a': 1,'b': 2,'c': 3,'d': 4,'e': 5,'f': 0}
    print(deep_min(data))

if __name__ == "__main__":
    main()
```
