```python
# Adoの誕生日
def ado_birthday(ado: str) -> None:
    p: int = 1
    n: int = len(ado)

    while n > 0:
        p *= 2
        n -= 1

    p_str = str(p)
    print(f'Adoの誕生日は{p_str[0:2]}月{p_str[2:4]}日')
    print(f'Happy Birthday {ado[0:3].capitalize()}ちゃん 🎂🎉')
    print(f'X.com/{ado[0:3]}{p_str}{ado[3:10]}')

def main() -> None:
    ado_birthday("adoimokenp")

if __name__ == "__main__":
    main()
```
