```python
# contextlib.suppress() - Context manager for suppressing exceptions
import os
from contextlib import suppress

with open("file1.txt", "w") as f:
    f.write("This is file is to test contextlib.suppress!")
    print("File1.txt created")

files_to_delete: list[str] = ["file1.txt", "file2.txt", "file3.txt"]

# Instead of using try-except block
for file in files_to_delete:
    with suppress(FileNotFoundError):
        os.remove(file)  # FileNotFoundError is suppressed
        print(f"{file} deleted")

# Useful for keyError fail
data = {"key1": "value1", "key2": "value2"}
with suppress(KeyError):
    del data["non_existent_key"]  # KeyError is suppressed

print(f"data after delete non_existent_key: {data}")
```
