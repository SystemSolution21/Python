```python
# memoryviews.ipynb
# memoryviews is a built-in type that provides a way to access the internal data of an object that supports the buffer protocol without copying.

# memoryview() - Create a memoryview object
import array

numbers: array.array[int] = array.array("i", [1, 2, 3, 4, 5])
view: memoryview = memoryview(numbers)
print(view)
print(view.tolist())
print(view[0])
```
