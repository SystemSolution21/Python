```python
# List comprehension 01
# get only last name from names list
names = ['Bobby C. Brown',
         'Chris Stevens',
         'Jacob G. Smith',
         'John Paul Davis']

def get_last_name(lst_names: list) -> list:
    last_name = [l_n.split(' ')[-1] for l_n in lst_names]
    return last_name

def main() -> None:
    print(get_last_name(names))

if __name__ == '__main__':
    main()
```

```python
# list Unpacking
inputs = ['John','Smith','United States','Blue','Brown','29']

sr_name, lt_name, *_, age = inputs
print(f'{sr_name} {lt_name} {age}')
```

```python
# list Initializing
from typing import Iterable, TypeVar

T = TypeVar('T')

def append_to_list(elements: Iterable[T], target: list[T] = None ) -> list[T]:
    if target is None:
        target = []

    target.extend(elements)
    return target


def main() -> None:
    num_lst = append_to_list([1, 2])
    print(num_lst)

    str_lst = append_to_list(['a', 'b'])
    print(str_lst)

    print(append_to_list([3, 4], num_lst))


if __name__ == "__main__":
    main()
```
