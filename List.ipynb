```python
# List comprehension 01
# get only last name from names list
names = ['Bobby C. Brown',
         'Chris Stevens',
         'Jacob G. Smith',
         'John Paul Davis']


def get_lst_name(lst_names: list) -> list:
    last_name = [l_n.split(' ')[-1] for l_n in lst_names]
    return last_name


def main() -> None:
    print(get_lst_name(names))


if __name__ == "__main__":
    main()
```

```python
from ast import Add
# list Unpacking
employee: list[str] = ['John','Smith','United States','Blue','Brown','29']

sur_name, lst_name, *_, age = employee
print(f'{sur_name} {lst_name} {age}')

# unpacking list using * operator
lst_num: list[int] = [1, 2, 3, 4]
a, *b, c = lst_num
print(f'{a=} {b=} {c=}')

def sum_num2(*args) -> int:
    return sum(args)

lst_num1: list[int] = [1, 2, 3]
lst_num2: list[int] = [4, 5]
lst_num3: list[int] = [6, 7, 8, 9]

print(sum_num2(*lst_num1, *lst_num2, *lst_num3))
```

```python
# list Initializing
from typing import Iterable, TypeVar

#T = TypeVar('T')

type = 'T'

def append_to_list(elements: Iterable[T], target: list[T] = None ) -> list[T]:
    if target is None:
        target = []

    target.extend(elements)
    return target


def main() -> None:
    num_lst = append_to_list([1, 2])
    print(num_lst)

    str_lst = append_to_list(['a', 'b'])
    print(str_lst)

    print(append_to_list([3, 4], num_lst))


if __name__ == "__main__":
    main()
```

```python
# copy list to a new list
num: list[int] = [1, 2, 3]
c_num: list[int] = num.copy()
c_num.append(4)
print(f'{id(num)=}')
print(f'{id(c_num)=}')
print(f'{num= }')
print(f'{c_num= }')

# deepcopy of reference type
from typing import Any
from copy import deepcopy

print("---- copy() alter reference ----")
a: list[Any] = [[1, 2],[3, 4]]
b:list[Any] = a.copy()
b[1][0] = "changed"
print(f'{a= }')
print(f'{b=} ')

print("---- deepcopy() do not alter reference ----")
a: list[Any] = [[1, 2], [3, 4]]
b: list[Any] = deepcopy(a)
b[1][0] = "changed"
print(f'{a= }')
print(f'{b=} ')
```
