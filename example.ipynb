```python
# Even Groups
''' Create a function, split_integer(), that takes two interger inputs "a" and "b".
    The "a" integer needs to be split into "b" groups as evenly as possible.
    Examples:
    split_integer(10, 4) -> [2, 2, 3, 3]
    split_integer(7, 3) -> [2, 2, 3]
    split_integer(3, 5) -> [0, 0, 1, 1, 1]
'''
def split_integer(number: int, parts: int) -> list[int]:

    # Step1: Divide number by parts to calculate quotient and remainder
    quotient, remainder = divmod(number, parts)

    # Step2: Count the number of parts that receive just the quotient
    base_parts_count = parts - remainder

    # Step3: Create array of integer where each elemet is quotient
    base_parts = base_parts_count * [quotient]

    # Step4: Create array of integer where each element is quotient plus one
    extra_parts = remainder * [quotient + 1]

    # Step5: Combine two arrays base_parts and extra_parts
    return base_parts + extra_parts

def test():
    tests = {
        "1:" : [2, 2, 3] == split_integer(7, 3),
        "2:" : [2, 2, 3, 3] == split_integer(10, 4),
        "3:" : [0, 0, 1, 1, 1] == split_integer(3, 5),
        "4:" : [9, 10] == split_integer(19, 2),
        "5:" : [1, 1, 2, 2] == split_integer(5, 4)
    }

    for k, v in tests.items():
        print(f'{k} {"passed" if v else "failed"}')

def main():
    test()
    print(split_integer(9, 6))

if __name__ == "__main__" :
    main()
```
