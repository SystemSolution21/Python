```python
from timeit import timeit

num_list = list(range(1_000_000))
num_set = set(range(1_000_000))

num_list_time = timeit("999_999 in num_list", number=1000, globals=globals())
num_set_time = timeit("999_999 in num_set", number=1000, globals=globals())

print(f'{num_list_time=: .6f} seconds')
print(f'{num_set_time=: .6f} seconds')

time_diff = ((num_list_time - num_set_time) / num_list_time) * 100
print(f'{time_diff=: .3f} %')
```
