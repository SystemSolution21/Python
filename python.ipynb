```python
# Adoの誕生日
def ado_birthday() -> None:
    p: int = 1
    n: int = 10

    while n > 0:
        p *= 2
        n -= 1

    print(f'{p = }')
    p_str = str(p)
    print(f'Adoの誕生日は{p_str[0:2]}月{p_str[2:4]}日です')

def main() -> None:
    ado_birthday()

if __name__ == '__main__':
    main()
```

```python
# Monetary amounts represent with integer rather than float
VAT: float = 0.18

PRICES: dict = {
    "burger" : 999,
    "fries" : 499,
    "drinks" : 199,
    "salad" : 1499
}

def order_food(price_lst: list) -> None:
    sub_total = sum(PRICES[item] for item in price_lst)
    vat = sub_total * VAT
    total = sub_total + vat
    print(f'Your order is ${total/100:.2f} (VAT ${vat/100:.2f})')

def main() -> None:

    order_food(["burger", "fries", "salad"])

if __name__ == "__main__":
    main()
```

```python
#swap variables
def swap() -> None:
    x: int = 3
    y: int = 4
    x, y = y, x
    print(x, y)

#unpack list
def unp_lst() -> None:
    num_lst: list = [1, 2, 3]
    x = y = z = int(0)
    x, y, z = num_lst
    print(x, y, z)

#return tuple
import random
def rtn_tpl() -> tuple[int, int]:
    x = random.randint(0, 100)
    y = random.randint(0, 100)
    return x, y

def main() -> None:
    swap()
    unp_lst()
    print(rtn_tpl())

if __name__ == "__main__":
    main()
```

```python
#List Comprehension
#squares
sqr_lst = [i*i for i in range(10)]
print(sqr_lst)
sqr_dict = {i: i*i for i in range(10)}
print(sqr_dict)

#cubes
def cube(i: int) -> int:
    return i**3

cub_lst = [cube(i) for i in range(10)]
print(cub_lst)

#if condition
if_lst = [i for i in range(20) if i%2==0]
print(if_lst)

#if else condition
if_els_lst = [0 if i<4 else i for i in range(10)]
print(if_els_lst)
```

```python
from traitlets.traitlets import Bool
#find prime numbers
def is_primes(n: int) -> bool:

    if n == 1:
        return False

    for d in range(2, n):
        if n % d == 0:
            return False
    return True

def main() -> None:

    for n in range(1, 6):
        print(n, is_primes(n))

if __name__ == "__main__":
    main()

import sympy
print(sympy.isprime(5))
print(list(sympy.primerange(1, 10)))
```

```python
# passing function as parameter to function
def print_length(f, g, x):
    return f(g(x))

print_length(print, len, "Hello World")

# random choice function
import random
def random_power():
    def f(x):
        return x** 2
    def g(x):
        return x** 3
    def h(x):
        return x** 4

    functions = [f, g, h]
    return random.choice(functions)

for _ in range(10):
    p = random_power()
    print(p(3))
```

```python
# Built-in __contains__ methods is useful to search in list

class Items:
    def __init__(self, items: list) -> None:
        self.items = items

    def __contains__(self, item):
        print(f'Searching for:')
        print(f'"{item}" in {self.items}')

        return item in self.items

def main() -> None:
    my_items = Items([1, 'a', 2, 3, 'd'])
    print('a' in my_items)
    print(5 in my_items)

if __name__ == '__main__':
    main()
```

```python
# prime factorization
def prime_factor(num: int) -> list:
    primes: list = []
    divisor: int = 2

    while num > 1:
        while num % divisor == 0:
            primes.append(num)
            num //= divisor
        divisor += 1
    return primes

def main() -> None:
    p_f = prime_factor(100)
    print(p_f)
    print(p_f[::-1]) # reverse list

if __name__ == '__main__':
    main()
```
