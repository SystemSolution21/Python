```python
#methodcaller
from operator import methodcaller

#filter
names: list[str] = ['Bob', 'James', 'Billy', 'Sandra', 'Blake']
starts_with_b: methodcaller = methodcaller('startswith', 'B')

filtered_by = filter(starts_with_b, names)
print(list(filtered_by))

#count
count_a: methodcaller = methodcaller('count', 'a')
sorted_by = sorted(names, key=count_a)
print(list(sorted_by))
```
