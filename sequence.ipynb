```python
# Executing c++ code on colab
%%writefile fibonacci.cpp
#include <iostream>
using namespace std;

void fibonacci(int num){
    int a = 0, b = 1, sum = 0;
    cout << "fibonacci sequence of " << num << ":" <<endl;
    while (a <= num){
        cout << a << " ";
        sum = a + b;
        a = b;
        b = sum;
    }
}

int main(){
    fibonacci(100);
    return 0;
}
```

```python
# fibonacci series
def fibonacci(n) -> list[int]:
    """Return a list of fibonacci series up to n."""
    result = []
    a, b = 0, 1
    while a <= n:
        result.append(a)
        a, b = b, a + b
    return result


def main() -> None:
    while (text := input("Input number to generate fibonacci series or 'q' to quit: "))!="q":
        if text.isdigit():
            n = int(text)
            f = fibonacci(n)
            print(f)
        else:
            print("Invalid Input!")


if __name__ == "__main__":
    main()
```

```python
import sys
from functools import wraps
from functools import cache

# memoization recursion
def memoize(func):
    cache = {}

    @wraps(func)
    def wrapper(*args, **kwargs):
        key = str(args) + str(kwargs)
        if key not in cache:
            cache[key] = func(*args, **kwargs)
        return cache[key]
    return wrapper

# Sum of fibonacci series
@memoize
def fibonacci(n) -> int:
    """Return sum of fibonacci series."""

    if n < 2:
        return n

    return fibonacci(n-1) + fibonacci(n-2)


'''
# memoization recursion using functools cache module
@cache
def fibonacci(n) -> int:
    """Return sum of fibonacci series."""

    if n < 2:
        return n

    return fibonacci(n-1) + fibonacci(n-2)
'''


def main() -> None:
    while(text := input("Input number to sum fibocacci series or 'q' to quit: ")) != "q":
        if text.isdigit():
            n = int(text)
            fib = fibonacci(n)
            print(f'{fib:,}')
        else:
            print("Invalid input!")


if __name__ == "__main__":
    sys.setrecursionlimit(10_000)
    main()
```
