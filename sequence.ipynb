```python
%%writefile fibonacci.cpp
#include <iostream>
using namespace std;

void fibonacci(int num){
    int a = 0, b = 1, sum = 0;
    cout << "fibonacci sequence of " << num << ":" <<endl;
    while (sum <= num){
//        cout << sum << endl;
        cout << sum << " ";
        sum = a + b;
        b = a;
        a = sum;
    }
}

int main(){
    fibonacci(10);
    return 0;
}
```

```python
# fibonacci series
def fibonacci(n) -> list[int]:
    """Return a list of fibonacci series up to n"""
    result = []
    a, b = 0, 1
    while a <= n:
        result.append(a)
        a, b = b, a + b
    return result


def main() -> None:
    while (num := input("Input number to generate fibonacci series or 'q' to quit: "))!='q':
        if num.isdigit():
            n = int(num)
            print(fibonacci(n))
        else:
            print("Invalid Input!")


if __name__ == '__main__':
    main()
```
