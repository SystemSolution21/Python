```python
%%writefile fibonacci.cpp
#include <iostream>
using namespace std;

void fibonacci(int num){
    int a = 0, b = 1, sum = 0;
    cout << "fibonacci sequence of " << num << ":" <<endl;
    while (sum <= num){
        cout << sum << " ";
        sum = a + b;
        b = a;
        a = sum;
    }
}

int main(){
    fibonacci(10);
    return 0;
}
```

```python
%%writefile fibonacci.cpp
#include <iostream>
using namespace std;

void fibonacci(int num)
{
    cout << "Input number: " << num << endl;
    cout << "Fibonacci sequence of "<< num << endl;

    int firstTerm = 0, secondTerm = 1, sumTerm = 0;

    while(firstTerm <= num)
    {
        cout << firstTerm << " ";
        sumTerm = firstTerm + secondTerm;
        firstTerm = secondTerm;
        secondTerm = sumTerm;
    }
}

int main()
{
    int num;
    cin >> num;
    fibonacci(num);
    return 0;
}
```

```python
# fibonacci series
def fibonacci(n) -> list[int]:
    """Return a list of fibonacci series up to n."""
    result = []
    a, b = 0, 1
    while a <= n:
        result.append(a)
        a, b = b, a + b
    return result


def main() -> None:
    while (text := input("Input number to generate fibonacci series or 'q' to quit: "))!="q":
        if text.isdigit():
            n = int(text)
            f = fibonacci(n)
            print(f)
        else:
            print("Invalid Input!")


if __name__ == "__main__":
    main()
```

```python
from functools import cache

# Sum of fibonacci series
@cache
def fibonacci(n) -> int:
    """Return sum of fibonacci series."""
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)


def main() -> None:
    while(text := input("Input number of items to generate sum of fibocacci series: ")) != "q":
        if text.isdigit():
            n = int(text)
            fib = fibonacci(n)
            print(f'{fib:,}')
        else:
            print("Invalid input!")


if __name__ == "__main__":
    main()
```

```python
# Generate sum of sequence of input number
def sum_numbers(n: int) -> int:
    return (n * (n + 1)) // 2

def main() -> None:
    while (text := input("Input number to generate sum of sequence or 'q' to quit:")) != "q":
        if text.isdigit():
            n = int(text)
            sum = sum_numbers(n)
            print(f'{sum: ,}')
        else:
            print("Invalid Input!")


if __name__ == "__main__":
    main()
```
