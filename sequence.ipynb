```python
%%writefile fibonacci.cpp
#include <iostream>
using namespace std;

void fibonacci(int num){
    int a = 0, b = 1, sum = 0;
    cout << "fibonacci sequence of " << num << ":" <<endl;
    while (sum <= num){
//        cout << sum << endl;
        cout << sum << " ";
        sum = a + b;
        b = a;
        a = sum;
    }
}

int main(){
    fibonacci(10);
    return 0;
}
```

```python
#1 generate fibonacci sequence using input() function
def fibonacci() -> None:
    num: int = 0

    print("Generate Fibonacci Sequence upto entered number!")
    while (num := input("Enter number or (type 'q' to quit)")) != 'q':
        a, b = 0, 1
        while a <= int(num):
            print(a, end=' ')
            a, b = b, a + b
        print()

def main() -> None:
    fibonacci()

if __name__ == '__main__':
    main()
```

```python
#fibonacci
def fibonacci() -> None:
    """Return a list containing fibonacci sequence up to entered number"""
    num: int = 0

    print("Generate Fibonacci Sequence up to entered number!")
    while (num := input("Please enter number or 'q' to quit: ")) != 'q':
        if num.isdigit():
            a, b = 0, 1
            result = []
            while a <= int(num):
                result.append(a)
                a, b = b, a + b
            print(result)
        else:
            print("Invalid input!")


def main() -> None:
    fibonacci()


if __name__ == '__main__':
    main()
```
