```python
#lambda expression to return a function
def multiple(n: int) -> int:
    return lambda x: x * n

double = multiple(2)
print(f'{double(5) = }')

triple = multiple(3)
print(f'{triple(5) = }')

#pass lambda expression as an keywords argument
pairs = [(1, 'one'), (2, 'two'), (3, 'three'), (4, 'four')]
pairs.sort(key = lambda pair: pair[1])
print(f'{pairs = }')
```
