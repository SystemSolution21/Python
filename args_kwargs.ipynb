```python
# *args and **kwargs
def order_pizza_hut(name, size, dough, *toppings, **details):
    print(f'Ordered {name} {size} {dough} pizza with the following toppings:')
    for topping in toppings:
        print(f'- {topping}')
    print("\nDetails of the order:")
    for key, value in details.items():
        print(f'- {key}: {value}')

def main() -> None:
    order_pizza_hut("マルゲリータ","L(12カット)", "ハンドトス(スタンダード生地)",
                    "ゴーダチーズ", "オニオン", delivery = True, tip = 0)

if __name__ == "__main__":
    main()
```

```python
# Default Arguments values
def ask_ok(prompt, retries=4, reminder="Please, try again!"):
    while True:
        reply = input(prompt)
        if reply in {"y", "ye", "yes"}:
            return True
        if reply in {"n", "no", "nop", "nope"}:
            return False
        retries -= 1
        if retries < 0:
            #raise ValueError("Invalid user response!")
            return "Invalid user response!"
        print(reminder)


def main() -> None:
    #result = ask_ok("Do you really want to quit? ")
    #result = ask_ok("Ok to overwrite the file? ", 2)
    result = ask_ok("Ok to overwrite the file? ", 2, "Come on, only yes or no")
    print(result)


if __name__ == "__main__":
    main()
```
