```python
# *args and **kwargs
def order_pizza(size, name, *toppings, **details):
    print(f'Ordered {size} {name} pizza with the following toppings:')
    for topping in toppings:
        print(f'- {topping}')
    print("\nDetails of the order are:")
    for key, value in details.items():
        print(f'- {key}: {value}')

def main():
    order_pizza("Large", "Margerata", "pepperoni", "olives", delivery=True, tip=5)

if __name__ == "__main__":
    main()
```
